```r
library(moments)
plot_sampling_distr <- function(n, iter){
    sample_means <- rep(NA, iter)

    for(i in 1:iter) {
        sample <- rchisq(n, df=1)
        sample_means[i] <- mean(sample)
    }
    print(paste("Skośność:", skewness(sample_means)))
    print(paste("Kurtoza: ", kurtosis(sample_means)))
    hist(sample_means, probability=TRUE)
    curve(dnorm(x, 1, sqrt(2/n)), 0, 2, add=TRUE)
}
```

```r
plot_sampling_distr(300, 10000)
```
